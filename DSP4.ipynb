```python
# merge all hadith csv file into a singular file
import pandas as pd

# Load both CSV files into DataFrames
df1 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_abidawud.csv')
df2 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_assalihin.csv')
df3 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_bukhari.csv')
df4 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_bulugh.csv')
df5 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_hisn.csv')
df6 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_ibnmajah.csv')
df7 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_malik.csv')
df8 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_mishkat.csv')
df9 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_mufrad.csv')
df10 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_muhammadiyah.csv')
df11= pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_muslim.csv')
df12 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_nasai.csv')
df13 = pd.read_csv('C:\\Users\\ACER\\OneDrive\\DSP 1 PROJECT\\HADITH DATASET DSP\\hadith_tirmidhi.csv')

# Combine the two DataFrames under the same table
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13], ignore_index=True)

# Save the combined DataFrame to a single CSV file
output_path = 'combined_hadith.csv'
combined_df.to_csv(output_path, index=False)

print(f"Combined data has been saved into a single CSV file: {output_path}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=data.Source.value_counts().values[:10], y=data.Source.value_counts().index[:10], palette="CMRmap")
plt.title('Distribution of Hadith Source')
plt.ylabel('Hadith Source')
plt.xlabel('Count')
plt.show()
```

```python
pip install --user wordcloud
```

```python
# Word Cloud for Hadith Texts
from wordcloud import WordCloud

all_text = " ".join(data['Hadith'].dropna().tolist())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Hadith Texts")
plt.show()
```

```python
# Length of Hadith Texts
data['Hadith Length'] = data['Hadith'].str.len()

# Distribution of Hadith Lengths

plt.figure(figsize=(10, 6))
sns.histplot(data['Hadith Length'], bins=30, kde=True, color="purple")
plt.title("Distribution of Hadith Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```
