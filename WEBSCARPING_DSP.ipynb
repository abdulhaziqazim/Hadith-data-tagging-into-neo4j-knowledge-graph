```python
# Base URL for Sahih al-Bukhari
base_url = "https://sunnah.com/bukhari/"

# Function to scrape chapter titles from the main Sahih al-Bukhari page
def scrape_chapter_titles(base_url):
    try:
        # Make a request to fetch the main page
        response = requests.get(base_url)
        response.raise_for_status()  # Raise an exception for HTTP errors

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all div elements with the class 'english english_book_name'
        chapter_divs = soup.find_all('div', class_='english english_book_name')

        # Extract chapter titles
        chapter_titles = [div.get_text(strip=True) for div in chapter_divs]

        return chapter_titles
    except Exception as e:
        print(f"An error occurred while scraping chapter titles: {e}")
        return []

# Function to scrape English Hadith content for a given chapter
def scrape_english_hadith(chapter_url):
    try:
        # Make a request to fetch the webpage
        response = requests.get(chapter_url)
        response.raise_for_status()  # Raise an exception for HTTP errors

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all elements with the class 'english_hadith_full'
        hadiths = soup.find_all('div', class_='english_hadith_full')

        # Extract the text content of each hadith
        hadith_texts = [hadith.get_text(strip=True) for hadith in hadiths]

        return hadith_texts
    except Exception as e:
        print(f"An error occurred while processing {chapter_url}: {e}")
        return []

# Scrape chapter titles
chapter_titles = scrape_chapter_titles(base_url)

# Scrape Hadiths for each chapter
all_hadiths = []
for chapter_num, chapter_title in enumerate(chapter_titles, start=1):
    chapter_url = f"{base_url}{chapter_num}"
    print(f"Scraping Hadiths for chapter: {chapter_title}")
    hadiths = scrape_english_hadith(chapter_url)
    for hadith in hadiths:
        all_hadiths.append(["Sahih Bukhari", chapter_title, hadith])  # Add "Sahih Bukhari" as the source

# Save the data to a CSV file
with open('hadith_bukhari.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Source', 'Book Title', 'Hadith'])  # Add a header row
    writer.writerows(all_hadiths)

print(f"Successfully scraped {len(all_hadiths)} hadiths with titles.")
```
